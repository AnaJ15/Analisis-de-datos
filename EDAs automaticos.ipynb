# EDA

Los EDA (E.D.A.: Exploratory data analysis) Análisis exploratorios de datos, son una forma de analizar los datos de forma estadística.

Fue definido por John W. Tukey

Consiste en una serie de descripciones de los datos a través de la llama Estadística descriptiva (media, mediana, moda, medidas de dispersión, desviación estándar, varianza y covarianza, etc.)

También se añaden comparaciones a través Estadística diferencial, con gráficas ya sean entre una o varias variables. La más importantes son las matrices de correlación, aunque también lo son los histogramas sobre distribuciones, boxplot de análisis de outliers, etc.

## Uso de los EDAs en Data Science

Aparte de lo "bonitos" que nos quedan los informes, un EDA es una pieza clave para saber cómo de **apto** es un conjunto de datos para ser utilizado por nuestros algoritmos de Aprendizaje Automático.

Si, porque el **99,9%** de las veces, los datasets no son "aptos" para realizar un proceso de AA directamente.

Y ese **0,1 %** de los que sí son aptos es porque ya han sido preparados para ello.

En definitiva:

- Un EDA tiene como objetivo examinar los datos en busca de distribuciones, valores atípicos y anomalías para dirigir las pruebas específicas de nuestra hipótesis.
- Un EDA presenta y resume los datos que se obtienen de una muestra, mediante técnicas de visualización y estadística.
- Un EDA nos ayuda a descubrir los patrones naturales y las características que son de mayor importancia.
- Un EDA nos permite preparar el informe que contiene los detalles de los conocimientos de datos que se pueden compartir con los propietarios de negocios y las partes interesadas.



### Los algoritmos y sus "delicadezas"

Un algoritmo de Aprendizaje Automático es como esa pieza de porcelana que al poco que la tocas te la cargas. Es decir, si queremos que funcione a pleno rendimiento, vamos a tener que hacer dos cosas:

    - Introducirle la información para el entrenamiento de la forma "adecuada"
    
    - Ajustarle los parámetros de cada algoritmo dependiendo de los datos introducidos
    
**¿Y cómo están los datos de forma adecuada? Algunas pistas:**

1. A estos "mimados" algoritmos **sólo le gustan los números**, nada de textos, que nunca han estudiado idiomas. 
    Para un algoritmo la diferenciación entre variables numéricas y categóricas no tiene sentido.
    
    - 1.a) Y más aún: A ser posible que no tengan una correlación numérica entre ellos, a no ser que eso indique algo a aprender.
    - 1.b) Más todavía: Si hay mucha diferencia entre los valores de distintas columnas, puede ser que ignore datos importantes
      Por lo que es mejor que todos los valores estén normalizados.
    - 1.c) Y por si fuera poco: Le encanta que le des todo de forma binaria o forma switch
    - 1.d) Para colmo: Le gustan las "campanas" (de Gauss y esas cosas)
    - 1.e) En relación a lo anterior: No le gustan mucho las excepciones ni los datos que van por libre.
    

2. Los valores **nulos se le indigestan** y provocan resultados impredecibles. Y ¡¡cuidado con los ceros!!
    - Aunque seguramente a quien se os van a indigestar es a vosotros cuando tengáis que resolver este punto.


3. Si hay muchos **valores distintos** en una columna, no sólo no sabe qué hacer con ellos, sino que además es posible que "aprenda" cosas sin sentido. 
    - Es decir, como las películas malas americanas, a los algoritmos le gusta que se le encasille todo en unas pocas categorías.
    
    
4. Si hay muchos **valores iguales** o "toda la columna tiene valores iguales" sólo le aporta digestiones lentas, pero nutrientes, pocos.


5. Si los datos están **relacionados temporalmente**, ni se los des. 
    - A los algoritmos les gusta barajear y esa afición a manejo de naipes no le sienta nada bien al tiempo.


6. Cuando le das **muchos datos, la digestión es lenta**, leeeennnnta, leeeeeeeeeeeennnnnnnnnntaaaaaaaaa
    - Por lo que nos tenemos que quedar con lo mínimo en cuanto a columnas y filas. 
    - Y mínimo es "MINIMO", sin contemplaciones. Si eliminas filas o columnas enteras, no te va a condenar la ONU ni ningún tribunal. (siempre que no pierdas información relevante para el aprendizaje)





#### Pues bien, los EDAs son los chivatos que nos van a "cantar" si a un dataset le pasa todo, parte o nada de lo anterior

Otra cosa será qué tenemos que hacer con los datos para resolver los 6 puntos anteriores.
